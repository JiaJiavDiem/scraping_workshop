```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Sun Jun  9 17:24:30 2019

@author: chris
"""

# This file contains the script to read the URLs from the SQLite database
# Start: SQLite database
# End: URLs in database

# Step2A: Read the URLs from the SQLite database
# Step2B: Scrape the URLs and write the content to the database
# Step2C: Retrieve URLs from content, write URLs to database
# Step2D: Use URLs to scrape, retrieve contents and write to database

import pandas as pd
import sqlite3
from sqlite3 import Error
import requests
import bs4
import html5lib
from lxml.html.soupparser import fromstring
```

```python
def create_connection(db_file):
    try:
        conn = sqlite3.connect(db_file)
        print("conn open")
        print(sqlite3.version)
        return conn
    except Error as e:
        print(e)

    return None
```

```python
def retrieve_data(sql, conn):
    df_data = pd.read_sql_query(sql, conn)
    return df_data
```

```python
def scrape(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.90 Safari/537.36'}
    res_1st = requests.get('http://' + url, headers=headers).text
    return res_1st
```

```python
def write_to_db(company, url, text, links_1st, id_link, conn):
    sql = """INSERT INTO firstScrape VALUES
            (
            ?,?,?,?,?
            )"""           
    list_values = [company, url, text, links_1st, id_link]
    print(list_values)        
    cur = conn.cursor()
    cur.execute(sql, list_values)
    conn.commit()
```

```python
def main():
    database = "scrape.db"
    table = "companies"
    sql = "SELECT * FROM " + table
    
    conn = create_connection(database)
    # Step 2A
    df_data = retrieve_data(sql, conn)
    
    # Step 2B
    for index, row in df_data.iterrows():
        id_link = index
        company = row['Company']
        url = row['Website']
        try:
            res_text = scrape(url)
        except:
            res_text = "No response"
                
        # Step 2C
        soup = bs4.BeautifulSoup(res_text, 'lxml')
        links = []
        for link in soup.find_all('a', href=True):
            print(link['href'])
            links.append(link['href'])
        links_1st = []
        for link in links:
            if "https" in link and not url in link:
                next
            elif link == url:
                next
            else:
                links_1st.append(link)
        links_1st = str(list(set(links_1st)))
        links_1st = links_1st.replace("'", "")
        links_1st = links_1st.replace('"', "")
        links_1st = links_1st.replace(" ", "")  
        write_to_db(company, url, res_text, links_1st, id_link, conn)    
    conn.close()
```

```python
if __name__ == "__main__":
    main()
```
